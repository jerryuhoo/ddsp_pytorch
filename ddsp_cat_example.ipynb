```python
import torch
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio
```

```python
from ddsp_pytorch.preprocess import preprocess
```

```python
model_path = "export/ddsp_cats_seg_pitchshift_pretrained.ts"
# you can also change the model to "ddsp_baseline_pretrained.ts" or "ddsp_cats_seg_pretrained.ts"
model = torch.jit.load(model_path)
```

```python
input_file = "violin_original.mp3"  # Replace with your input file
x, pitch, loudness = preprocess(input_file, 16000, 160, 64000, False)
pitch = pitch[0]
loudness = loudness[0]
pitch = torch.tensor(pitch).unsqueeze(0).unsqueeze(-1).float()
loudness = torch.tensor(loudness).unsqueeze(0).unsqueeze(-1).float()
```

```python
audio = model(pitch, loudness)
```

```python
audio_numpy = audio[0].detach().cpu().numpy()
audio_16bit = np.int16(audio_numpy * 32767)
write("test_output.wav", 16000, audio_16bit)
```
